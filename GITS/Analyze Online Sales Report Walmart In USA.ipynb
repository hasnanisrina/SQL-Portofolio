# Change Wrong Format

```python
# Data type for order_date and Customer Since is Date
df['order_date'] = pd.to_datetime(df['order_date'])
df['Customer Since'] = pd.to_datetime(df['Customer Since'])
df.head()
```

```python
df.info()
```

# Check missing value

```python
# Check number of missing value in each column
df.isnull().sum()
```

```python
# Replace missing value in month and year from order_date
df['month'] = df['month'].fillna(pd.to_datetime(df['order_date'], format='%m-%y'))
df['year'] = df['year'].fillna(pd.to_datetime(df['order_date'], format='%Y'))
```

```python
# Check number of missing value in each column
df.isnull().sum()
```

```python
# Replcae middle initial and cust_id
df['Middle Initial'].fillna(method = 'ffill', inplace = True)
df['cust_id'].fillna(method = 'ffill', inplace = True)
```

```python
# Check number of missing value in each column
df.isnull().sum()
```

```python
# function for price, qty_ordered, total, value, discount_amount
price = df["value"] / (df["qty_ordered"]-1)
qty_ordered = (df["value"]/df["price"]) + 1
total = (df["qty_ordered"]-1) * df["price"] - df["discount_amount"]
value = (df["qty_ordered"]-1) * df["price"]
disc_amount = (df["qty_ordered"]-1) * df["price"] * df["Discount_Percent"]
```

```python
# Replace price, qty_ordered, total, value, discount_amount
df["price"].fillna(price, inplace = True)
df["qty_ordered"].fillna(qty_ordered, inplace = True)
df["total"].fillna(total, inplace = True)
df["value"].fillna(value, inplace = True)
df["discount_amount"].fillna(disc_amount, inplace = True)
```

```python
# Check number of missing value in each column
df.isnull().sum()
```

```python
#Remove missing data
df.dropna(inplace = True)
```

```python
# Check number of missing value in each column
df.isnull().sum()
```

```python
# Check shape of data
df.shape
```

# Treatment Data Outlier

```python
df = df.astype({"qty_ordered":'int64'}) 
df.dtypes
```

```python
df['qty_ordered'] = df['qty_ordered'] - 1
```

```python
df['qty_ordered'].sort_values()
```

```python
df.corr()
```

```python
numerical_features = list(df.select_dtypes(include=['int64','float64']).columns)
# del numerical_features[-3] #delete target/label value, menyesuaikan dengan lokasi column target
numerical_features
```

```python
label = 'status'
```

```python
# Boxplot to check outlier
# Ignore item_id, cust_id,ref_num, zip
for col in numerical_features:
    df.boxplot(column=col, by=label, figsize=(12,6))
    plt.title(col)
    
plt.show()
```

```python
# qty_ordered : kalau mau hapus data diatas 300 (atau di atas 100 ajaa? beauty and grooming, superstore category ada 54 data )
# price : kalau mau hapus data di atas 40.000 (beauty and entertainment category, ini kayanya typo jd its okay kalau dihapus)
# value : data di atas 60.000 di hapus aja 
# disc amount (harga diskon) : hapus di atas 15.000
# Total : hapus di atas 60.000 (kayanya samain ke value, soalnya value itu sama kaya total
#         cuman value itu belum dikurangi discount)
# age aman gaada outlier
# discount percent > 70
```

```python
# Remove outlier
df = df[df['qty_ordered'] <= 300]
df = df[df['price'] <= 40000]
df = df[df['value'] <= 60000]
df = df[df['discount_amount'] <= 15000]
df = df[df['total'] <= 60000]
df = df[df['Discount_Percent'] <= 70]
```

```python
# Boxplot to check outlier
# Ignore item_id, cust_id,ref_num, zip
for col in numerical_features:
    df.boxplot(column=col, by=label, figsize=(12,6))
    plt.title(col)
    
plt.show()
```

# Preparation for Dashboard

```python
# Mengelompokkan umur
def categorize(x):
  if x <= 25:
    return "<26"
  elif x > 25  and x <= 35:
    return "26-35"
  elif x > 35 and x <=45:
    return "36-45"
  elif x > 45 and x <= 55:
    return "46-55"
  elif x > 55 and x <= 65:
    return "56-65"
  elif x > 65:
    return ">65"
 

df["range_age"] = df["age"].apply(lambda x: categorize(x))
df.head()
```

```python
# Lamanya menjadi customer
df['customer_days'] = df['order_date'].sub(df['Customer Since'], axis = 0)
df['customer_years'] = (df['customer_days']/np.timedelta64(1, 'Y')).astype('int')
df.head()
```
